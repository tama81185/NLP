```python
#create a function to remove stop words.
STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
    return " ".join([word for word in text.split() if word not in STOPWORDS])
```

```python
#Removal of Frequent Words
from collections import Counter
```

```python
word_count = Counter()
for text in df['review']:
    for word in text.split():
        word_count[word] += 1

word_count.most_common(10)
```

```python
FREQUENT_WORDS = set(word for (word, wc) in word_count.most_common(3))
def remove_freq_words(text):
    return " ".join([word for word in text.split() if word not in FREQUENT_WORDS])
```

```python
df['review']=df['review'].apply(remove_freq_words)
df.head()
```

```python
#Removal of Rare Words

RARE_WORDS = set(word for (word, wc) in word_count.most_common()[:-10:-1])
RARE_WORDS
```

```python
def remove_rare_words(text):
    return " ".join([word for word in text.split() if word not in RARE_WORDS])
```

```python
df['review']=df['review'].apply(remove_rare_words)
df.head()
```

```python
df['review']=df['review'].apply(remove_spl_chars)
df.head()
```

```python
#Chat word treatment


def load_slang_dict(filename):
    slang_dict = {}
    with open(filename, 'r', encoding='utf-8') as file:
        for line in file:
            if '=' in line:  # Handling lines with '=' separator
                key, value = line.strip().split('=', 1)
            elif ':' in line:  # Handling lines with ':' separator
                key, value = line.strip().split(':', 1)
            else:
                continue  # Skip lines that don't follow the format
            slang_dict[key.strip()] = value.strip()
    return slang_dict

# Usage example
filename = "slang.txt"
slang_dict = load_slang_dict(filename)
print(slang_dict)
```

```python
chat_words = load_slang_dict("slang.txt")
```

```python
def chat_conversion(text):
    """Replaces slang words with their full meanings."""
    new_text = []
    for w in text.split():
        if w.upper() in chat_words:
            new_text.append(chat_words[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)
```

```python
chat_conversion('IMHO he is the best')
```

```python
#Manually Create a Bangla Stopwords List
bangla_stopwords = set([
    "আমি", "আমার", "এটি", "এই", "তুমি", "তোমার", "কেন", "যদি", "কোথায়", "সে",
    "তারা", "যারা", "কি", "কখন", "হয়", "না", "হলো", "থেকে", "জন্য", "সেই",
    "যা", "যেখানে", "কিভাবে", "কি", "এবং", "এখন", "তাদের", "ছিল", "করছে",
    "করতে", "করবেন", "নেই", "হয়েছে", "বেশি"
])
```

```python
import re

def remove_bangla_stopwords(text):
    words = re.findall(r'\b\w+\b', text)  # Tokenize text
    filtered_words = [word for word in words if word not in bangla_stopwords]
    return " ".join(filtered_words)

# Example usage
bangla_text = "আমি তোমার জন্য অপেক্ষা করছিলাম কিন্তু তুমি এলো না।"
clean_text = remove_bangla_stopwords(bangla_text)
print(clean_text)
```

```python
def remove_stopwords(text):
    new_text = []

    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    x = new_text[:]
    new_text.clear()
    return " ".join(x)
```

```python
import re
def remove_emoji(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
```

```python
#Stemming
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
def stem_words(text):
    return " ".join([ps.stem(word) for word in text.split()])
```

```python
##Stemming
import nltk
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

sentence = "He was running and eating at same time. He has bad habit of swimming after playing long hours in the Sun."
punctuations="?:!.,;"
sentence_words = nltk.word_tokenize(sentence)
for word in sentence_words:
    if word in punctuations:
        sentence_words.remove(word)

sentence_words
print("{0:20}{1:20}".format("Word","Lemma"))
for word in sentence_words:
    print ("{0:20}{1:20}".format(word,wordnet_lemmatizer.lemmatize(word,pos='v')))
```
